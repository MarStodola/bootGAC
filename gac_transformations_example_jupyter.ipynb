```python
from bootGAC import *
from clifford.tools.g3c import *
```

```python
ell = init_conic(center_x=2, center_y=3,
                 axis_a=5, axis_b=3, theta=-1/3 * np.pi)
hyp = init_conic(center_x=2, center_y=3,
                 axis_a=1, axis_b=-4/3, theta=-1/6 * np.pi)
par = init_conic(center_x=2, center_y=3, axis_a=2, axis_b=0, theta=1/6 * np.pi)
```

```python
rot, rot_inv = rotor_and_rotor_inv_gac(np.pi / 2.5)
tr, tr_inv = translator_and_translator_inv_gac(2, 7)
hyp_rot = np.matmul(rot, np.matmul(hyp, rot_inv))
ell_rot = np.matmul(rot, np.matmul(ell, rot_inv))
par_rot = np.matmul(rot, np.matmul(par, rot_inv))

hyp_tr = np.matmul(tr, np.matmul(hyp, tr_inv))
ell_tr = np.matmul(tr, np.matmul(ell, tr_inv))
par_tr = np.matmul(tr, np.matmul(par, tr_inv))
```

```python
sc_rot = GanjaScene()
add_conic_to_scene(sc_rot, ell, Color.BLUE)
add_conic_to_scene(sc_rot, ell_rot, Color.BLACK)
# add_conic_to_scene(sc_rot, hyp, Color.CYAN)
# add_conic_to_scene(sc_rot, hyp_rot, Color.MAGENTA)
add_conic_to_scene(sc_rot, par, Color.CYAN)
add_conic_to_scene(sc_rot, par_rot, Color.MAGENTA)
draw(sc_rot, scale=0.1)
```
